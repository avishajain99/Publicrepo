## Data Visualization

Plotting line graphs to show the populations of India and China over time

```python
## Choose two specific countries : India and China
x = 'India'
y = 'China'

## Filter data for the selected countries
x_data = total[total['Country Name'] == x]
y_data = total[total['Country Name'] == y]
```

```python
## Extract relevant columns for plotting (excluding 'Country Name' column)
x_years = x_data.columns[1:]
y_years = y_data.columns[1:]
```

```python
## Plotting line graph for showing the populations of India over time
plt.figure(figsize = (18,6))
plt.subplot(2,1,1)
plt.plot(x_years,x_data.iloc[:, 1:].values.flatten(),marker = 'o',label = x)
plt.title(f'Total Population of India over time')
plt.xlabel('Years')
plt.ylabel('Population')
plt.legend()
plt.xticks(x_years, rotation=90)
plt.show()

## Plotting line graph for showing the populations of China over time
plt.figure(figsize = (18,6))
plt.subplot(2,1,2)
plt.plot(y_years,y_data.iloc[:,1:].values.flatten(),marker = 'o',label = y)
plt.title(f'Total Population of China over time')
plt.xlabel('Years')
plt.ylabel('Population')
plt.legend()
plt.xticks(y_years, rotation=90)

#plt.tight_layout()
plt.show()
```

```python
# Create the bar graph
fig, (axis1) = plt.subplots(1, figsize = (33,12))
sns.barplot(x = x_years, y = x_data.iloc[:,1:].values.flatten(), ax= axis1)

# Add title and axis labels
plt.title(f'Bar Graph showing population of India over time', fontsize = 16)
plt.xlabel('Year')
plt.ylabel('Population')
#plt.xticks(x_years, rotation=90)

# Display the plot
plt.show()
```

```python
## Selecting a specific region to visualize

selected_region = 'World'
```

```python
## Filtering data for the selected region
selected_reg_total = total[total['Country Name'] == selected_region]
selected_reg_male = male[male['Country Name'] == selected_region]
selected_reg_female = female[female['Country Name'] == selected_region]
```

```python
## Extract relevant columns for plotting (excluding 'Country Name' column)
years = selected_reg_total.columns[1:]
```

```python
## Plotting a STACKED bar chart showing World's population of male and female

# Create the bar graph
plt.figure(figsize = (33,12))
plt.bar(years, selected_reg_male.iloc[0,1:], label = 'Male', color = 'steelblue', alpha = 0.7)
plt.bar(years, selected_reg_female.iloc[0,1:], label = 'Female', bottom = selected_reg_male.iloc[0,1:], color = 'red', alpha = 0.7)

# Add title and axis labels
plt.title(f'Population distribution in {selected_region} by Gender (1960-2022)', fontsize = 16)
plt.xlabel('Year', fontsize = 10)
plt.ylabel('Population', fontsize = 10)
plt.xticks(rotation=90)
plt.legend()

# Display the plot
plt.show()
```

### Sorting data based on highest to lowest populations for 2022

```python
total_population_sort = total.sort_values(by = '2022', ascending = False)
total_population_sort
```

## Top 10 regions with the highest population in 2022

```python
top_ten = total_population_sort.head(10)
print('Top 10 regions with the most population')
print(top_ten)
```

## Top 10 regions with the highest population in the year 1988 and 2022

```python
## Creating a bar graph
fig, [ax1, ax2] = plt.subplots(2,1,figsize = (20, 22))
sns.barplot(x = '1988', y = 'Country Name', data = top_ten, ax = ax1).set(title= 'Top 10 regions with most population in 1988')
# plt.title(fontsize = 13)

sns.barplot(x = '2022', y = 'Country Name', data = top_ten, ax = ax2).set(title = 'Top 10 regions with most population in 2022')
# plt.title(fontsize = 13)

# renaming the axes
ax1.set(xlabel="Total population", ylabel="Region")
ax2.set(xlabel="Total population", ylabel="Region")

## Changing default axes color
plt.rcParams['axes.facecolor'] == 'black'

plt.show()
```
